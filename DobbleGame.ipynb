```python
from random import choice
import emoji

# class DobbleDeck contains the deck of cards, and all methods for manipulating
# and showing
class DobbleDeck:
    # contructor loads creates a card list and loads emojis for the cards.
    # it also checks the validity calling that function
    # and also builds list of DobbleCard
    def __init__(self):
        self.card_list = []
        self.imageDict = dict()
        fin = open('emoji_names.txt',"r")
        lines = fin.readlines()
        for i, el in enumerate(lines):
            self.imageDict[i+1] = emoji.emojize(el.strip())
        deck_dictionary = self.create_value_dictionary()
        self.check_validity(deck_dictionary)
        for key, card_image_set in deck_dictionary.items():
            self.add_card(card_image_set)

    # adds a DobbleCard to a list
    def add_card(self, card_images_set):
        dobble_card = DobbleCard(card_images_set)
        self.card_list.append(dobble_card)

    # chooses a random card and removes it from the list
    def remove_card(self):
        card = choice(self.card_list)
        self.card_list.remove(card)
        return card

    # displays a card by printing to console
    def play_card(self, card):
        show_card_string = ''
        for i in card.card_images_set:
            show_card_string += self.imageDict[i+1]
        print(show_card_string)

    # code provided by exercise to generate DobbleCards
    def create_value_dictionary(self):
        nIm = 8
        n = nIm - 1
        r = range(n)
        rp1 = range(n+1)
        c = 0

        deck_dictionary = {}
        # First card
        c += 1
        card_one = 'Card ' + str(c)

        images_set = set([])

        for i in rp1:
            images_set.add(i+1)

        deck_dictionary[str(c)] = images_set

        images_set2 = set([])
        # n following cards
        for j in r:
            c = c+1
            images_set2.add(1)
            for k in r:
                images_set2.add(n+2 + n*j +k)

            deck_dictionary[str(c)] = images_set2
            images_set2 = set([])
        # n x n following cards

        images_set3 = set([])
        for i in r:
            for j in r:
                c = c+1
                images_set3.add(i+2)
                for k in r:
                    images_set3.add((n+1 +n*k + (i*k+j) % n)+1)
                deck_dictionary[str(c)] = images_set3
                images_set3 = set([])

        return deck_dictionary

    # checks the validity of the cards and throws exception if not valid
    def check_validity(self, deck, verbose=False):
        for key1, values1 in deck.items():
            for key2, values2 in deck.items():
                if key1 != key2:
                    intersect = values1.intersection(values2)
                    is_pair_valid = len(intersect) == 1

                    if verbose:
                        print("Comparing card", key1, "to card", key2, ":")
                        print("The cards have", intersect, "in common")
                        if is_pair_valid:
                            print("This pair is a valid pair")
                        else:
                            print("This pair is not a valid pair")

                    if not is_pair_valid:
                        raise Exception(key1, key2,  " is not a valid pair", intersect)
```

```python
# stores card images for specific card
class DobbleCard:
    def __init__(self, card_images_set):
        self.card_images_set = card_images_set

```

```python
# main logic flow of game
def play_dobble_game():

    dobble_deck = DobbleDeck()


    number_cards_play = -1
    while number_cards_play < 1 or number_cards_play > 56:
        try:
            number_cards_play = int(input("How many cards (<56)?"))
        except ValueError:
            number_cards_play = -1

    a, b = 0, 0
    first_card, second_card = None, None
    print("If you want to record a draw type 'd' or 'D'")
    print()

    for i in range(number_cards_play):
        if first_card is None:
            first_card = dobble_deck.remove_card()

        dobble_deck.play_card(first_card)
        print()

        second_card = dobble_deck.remove_card()
        dobble_deck.play_card(second_card)

        winner = input("Who wins, A or B?")
        winner = winner.lower()

        while winner not in ('a','b','d'):
            winner = input("Who wins, A or B? (d for draw)")

        if winner == "a":
            a += 1
        elif winner == "b":
            b += 1
        print("Score A: ", a, "Score B: ", b)

        first_card = second_card

    print("Final Score A: ", a, "Score B: ", b)
```

```python
play_dobble_game()
```
